```python
#creat dataframe of versicolor petal
versicolor_petal_df= iris_df[iris_df.iris_class == 'Iris-versicolor']
```

```python
versicolor_petal_df
```

```python
#Create an array for versicolor_petal length and width and provide sample variance and standard deviation
versicolor_petal_length_arr=np.array(versicolor_petal_df['petal_length_cm'])
versicolor_petal_width_arr=np.array(versicolor_petal_df['petal_width_cm'])
```

```python
variance_length = np.var(versicolor_petal_length_arr)
```

```python
variance_length
```

```python
variance_width = np.var(versicolor_petal_width_arr)
```

```python
variance_width
```

```python
standard_daviation_length=np.std(versicolor_petal_length_arr)
```

```python
standard_daviation_length
```

```python
standard_daviation_width=np.std(versicolor_petal_width_arr)
```

```python
standard_daviation_width
```

```python
#Make a scatter plot between petal length and width
scatter_plot = plt.plot(versicolor_petal_length_arr, versicolor_petal_width_arr,marker='.', linestyle='')
```

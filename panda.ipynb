```python
import pandas as pd
import numpy as np
```

```python
t = np.arange(0,10,0.1)

# the angle(theita)=omega*time

omega = 2*np.pi*1.5

y = np.sin(omega*t)
z = np.cos(omega*t)

#label the x,y,z columns:
df = pd.DataFrame({'Time':t, 'YVals':y, 'ZVals':z})
```

```python
#Extract the data in lines 10–15 from “Yvals” and “ZVals”, and write them to the file “out.txt”S
extractData_file = 'out.txt'
df[10:16][['YVals', 'ZVals']].to_csv(extractData_file)
```

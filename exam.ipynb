```python
#FPL Valuation

dfview = epldata_df.nlargest(10, 'fpl_value').sort_values('fpl_value',ascending=False)
plt.subplots(figsize=(15,6))
sns.barplot(x="name", y="fpl_value",data=dfview ,palette='hot',edgecolor=sns.color_palette('dark',7))
plt.xticks()
plt.show()
```

```python
#Whoâ€™s stocking up at which position?

pd.crosstab(epldata_df.club,epldata_df.position).plot.bar(stacked=True,width=1,color=sns.color_palette())

fig=plt.gcf()
fig.set_size_inches(12,8)
plt.title('Number Of Player in every Clubs and position')

plt.show()
```
